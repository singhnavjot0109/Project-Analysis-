```python
import pandas as pd
xl = pd.ExcelFile("/content/sample_data/TableauSalesData.xlsx")
SalesData = xl.parse("Orders")
```

```python
JustTables = SalesData
```

```python
Tabsreg = JustTables[["Sub-Category","Customer Name", "Profit", "Sales"]]
SubCatcust = Tabsreg.loc[Tabsreg["Customer Name"]=="Sub-Category"]
SubCatcust= SubCatcust.groupby(by="Sub-Category").sum().sort_values(by="Profit",ascending = True)
display(SubCatcust)
```

```python
TableRegion = JustTables[["Region","Profit","Sales"]]
SegTotalProfSales = TableRegion.groupby(by="Region").sum().sort_values(by="Profit",ascending = False)
display(SegTotalProfSales)
```

```python
Subcatssale = JustTables[["Sub-Category","Sales","Profit"]]
artsaletable = Subcatssale.loc[Subcatssale["Sub-Category"]=="Art"]
SegSubcatsale= artsaletable.groupby(by="Sub-Category").sum().sort_values(by = "Sales")
display(SegSubcatsale)
```

```python
Artprod = JustTables[["Sub-Category","Product Name","Sales","Profit"]]
artitem = Artprod.loc[Artprod["Sub-Category"] == "Art"]
SegArtprods = artitem.groupby(by = "Product Name").sum().sort_values(by = "Profit")
display(SegArtprods)
```

```python
TableRegion = JustTables[["Region","State","Profit","Sales"]]
WestTable = TableRegion.loc[TableRegion["Region"]=="West"]
WestProfSales = WestTable.groupby(by="State").sum().sort_values(by = "Profit",ascending = False)
display(WestProfSales)
```
